#### Using embedded feature selection technique ####

```python
attributes = [col for col in wine.columns if col != "Wine"]

# partition: train/test = 70/30
train_x, test_x, train_y, test_y = train_test_split(wine[attributes], wine["Wine"], test_size=0.2, random_state=123)

# convert numpy arrays to data frames
df_train_x = pd.DataFrame(train_x, columns=attributes)
df_test_x = pd.DataFrame(test_x, columns=attributes)
df_train_y = pd.DataFrame(train_y, columns=["Wine"])
df_test_y = pd.DataFrame(test_y, columns=["Wine"])

print ("shapes")
print (df_train_x.shape)
print (df_test_x.shape)
print (df_train_y.shape)
print (df_test_y.shape)

print 
attributes
print ("Wine counts")
print (wine["Wine"].value_counts())
print (df_train_y["Wine"].value_counts())
print (df_test_y["Wine"].value_counts())
```

```python
f_val, p_val = chi2(df_train_x, df_train_y["Wine"]) 

# print the Chi-squared valus and p values
df_scores = pd.DataFrame(zip(attributes, f_val, p_val), columns=["feature", "chi2", "p"])
df_scores["chi2"] = df_scores["chi2"].round(2)
df_scores["p"] = df_scores["p"].round(3)
#print df_scores.sort_values("chi2", ascending=False)

# use features with p < 0.05
sel_cols = df_scores[df_scores["p"]<0.10]["feature"].values
print ("\nSelected features: %d" % len(sel_cols))
print (sel_cols)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(df_train_x[sel_cols], train_y)
pred_y = dtc.predict(df_test_x[sel_cols])

print ("f1:{:.2f}%".format(f1_score(pred_y, test_y, average="micro")))
```
